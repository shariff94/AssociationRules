```python
import pandas as pd
```

```python
# Read the CSV
data = pd.read_csv("store_data.csv",header=None)
data.head()
```

```python
data.shape
```

```python
records={}
for i,row in data.iterrows():
    records[i] = list(row[row.notna()])
```

```python
from apyori import apriori
```

```python
rule_set = apriori(list(records.values()),min_support=0.004, min_confidence=0.2,min_lift=2)
```

```python
association_results = list(rule_set)
```

```python
print(len(association_results))
```

```python
for i in association_results:
    print(i.items)
    print(i.ordered_statistics)
    print(i.support)
    print("#########################################################################################")
```
